```python
import pandas as pd
import numpy as np 
import folium
pd.set_option('display.max_columns', None)
data = pd.read_csv('Data-Collisions.csv')
```

```python
df = data.filter(items= ['SEVERITYCODE', 'X', 'Y', 'INCKEY', 'ADDRTYPE', 'LOCATION', 'SEVERITYDESC', 'COLLISIONTYPE',
       'PERSONCOUNT', 'PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT', 'INCDATE',
       'INCDTTM', 'JUNCTIONTYPE','INATTENTIONIND', 'UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND',
       'PEDROWNOTGRNT', 'SPEEDING', 'HITPARKEDCAR',  'ST_COLCODE','SDOT_COLDESC'])
df.head(1)
```

```python
df.shape
```

```python
#include only rows with X, Y coordinates
df1 = df.dropna(subset=['X', 'Y'])
df1.head(3)
```

```python
df_coord = df.dropna(subset=['X', 'Y', 'SEVERITYCODE'])
df_coord.shape
```

```python
#incident count by coordinates
df_unique_coords = df_coord.groupby(['X', 'Y']).count()
#Cols X and Y are now indices
df_unique_coords.shape
```

```python
sorted_top_500 = df_unique_coords.sort_values(by='SEVERITYDESC', ascending=False).iloc[:500]
top = sorted_top_500.reset_index()
top.shape
```

```python
#top 100 locations with highest incidents
num = 50
top_locs = top.iloc[:num]
top_locs.reset_index(inplace=True)
top_locs.tail()
```

```python
#bottom 100 locations with highest incidents
start_num = 475
bottom_locs = top.iloc[start_num:500]
bottom_locs.tail()
```

```python
#severity type by coordinates
df_type = df_coord.groupby(['X', 'Y', 'SEVERITYDESC']).agg({'INCKEY': 'count'})
df_type.reset_index(inplace=True)
df_type.tail()
```

```python
#sort to get top
top_num = 100
type_top = df_type.sort_values(by='INCKEY', ascending=False).iloc[:top_num]
```

```python
#split btw severity
df_1 = df_type[df_type['SEVERITYDESC'] == 'Property Damage Only Collision']
df_1.head()
```

```python
#top severity level 1
top100_l1 = df_1.sort_values(by='INCKEY', ascending=False).iloc[:50]
top100_l1.reset_index(inplace=True)
top100_l1.head()
```

```python
#split btw severity
df_2 = df_type[df_type['SEVERITYDESC'] == 'Injury Collision']
df_2
```

```python
#top severity level 2
top100_l2 = df_2.sort_values(by='INCKEY', ascending=False).iloc[:50]
top100_l2.reset_index(inplace=True)
top100_l2.head()
```

```python
limit = 10000
df_incidents = df_type.iloc[0:limit, :]
```

```python
#Seattle coordinates
latitude = 47.608013
longitude = -122.335167
```

```python
seattle_map = folium.Map(location=[latitude, longitude], zoom_start=12)
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, severity, in zip(type_top.Y, type_top.X, type_top.SEVERITYDESC):
    if severity == 'Property Damage Only Collision':
        incidents.add_child(
            folium.features.CircleMarker(
                [lat, lng],
                radius=5, # define how big you want the circle markers to be
                color='yellow',
                fill=True,
                fill_color='blue',
                fill_opacity=0.6
            )
        )
    else:
        incidents.add_child(
            folium.features.CircleMarker(
                [lat, lng],
                radius=5, # define how big you want the circle markers to be
                color='orange',
                fill=True,
                fill_color='red',
                fill_opacity=0.6
            )
    )
        

# add incidents to map
seattle_map.add_child(incidents)
```
